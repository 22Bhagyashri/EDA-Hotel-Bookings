# Data Wrangling

```python
# replacing null values in children columns with 0 assuming that family has 0 childre
# replacing null values in company and agent columns with 0 assuming those rooms were booked without complany agent

df['children'].fillna(0,inplace=True)
df['company'].fillna(0,inplace=True)
df['agent'].fillna(0,inplace=True)


# replace null values in contry columns  as others
df['country'].fillna('others',inplace=True)
```

```python
# checking null values after replacing 

df.isnull().sum()
```

```python
# drop the company column as 

df.drop(['company'],axis=1)
```

```python
# adding a nue columns to make our data analysis ready

df['total_pepole']=df['adults']+df['babies']+df['children']

df['total_stay']=df['stays_in_weekend_nights']+df['stays_in_week_nights']
```

# Visualization code

# Creating a pie chart

```python
hotel_count = df.hotel.value_counts()

hotel_count.plot.pie(figsize=(9,7), autopct = '%1.2f%% ', shadow=True,fontsize =15, startangle=50)

plt.title('Hotel Booking Percentage')
plt.axis('equal')
plt.show()
```

```python
#storing hotel cancelation
canceled_hotel = df.is_canceled.value_counts()

#create a pie chart
canceled_hotel.plot.pie(figsize=(9,7), explode= (0.05,0.05), autopct = '%1.2f%% ', shadow=True,fontsize =15, startangle=50)

#title pie chart
plt.title('Hotel Booking cancelation percentage')
plt.axis('equal')
plt.show()
```

# Which type of meal ia most preffered by guests ?


```python
# count meal type
meal_count = df.meal.value_counts()

meal_name =df ['meal']. unique()

# creating dataset of each meal type and count
meal_df = pd.DataFrame(zip(meal_name,meal_count),columns =['meal_name','meal_count'])

# Value of bar chart
plt.figure(figsize=(15,5))
g=sns.barplot(data=meal_df , x='meal_name' , y= 'meal_count')
g.set_xticklabels(meal_df['meal_name'])
plt.title("most meal preffered type", fontsize= 30)
plt.show()
```

# Which year has the most bookings?

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df['arrival_date_year'],hue=df['hotel'])
plt.title("Number of booking across year")
plt.show()
```

# Which month has the most booking in each hotel type ?

```python
plt.figure(figsize=(13,5))
sns.countplot(x=df['arrival_date_month'],hue=df['hotel'])
plt.title("Number of booking across months",fontsize=25)
plt.show()
```

# From which country most guests come?

```python
country_df =df['country'].value_counts().reset_index().rename(columns={'index' :'country','country':'guests count'})[:10]

plt.figure(figsize=(13,4))
sns.barplot(x=country_df['country'],y=country_df['guests count'])
plt.title("Number of guests from each country",fontsize=25)
plt.show()
```

# Which room type is most preffered by guests ? 


```python
plt.figure(figsize=(10,5))
sns.countplot(x=df['reserved_room_type'],order=df['reserved_room_type'].value_counts().index)
plt.title("Prefferd room type by guests")
plt.show()
```

# Which room type is most assigned?

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df['assigned_room_type'],order=df['assigned_room_type'].value_counts().index)
plt.title("Assigned room type by guests")
plt.show()
```

# Which is the percentage of repeated guests?

# How long pepole stay in the hotel?

```python
not_cancelled = df[df['is_canceled']==0]
hotel_stay= not_cancelled[not_cancelled['total_stay']<=15]

plt.figure(figsize=(15,5))
sns.countplot (x= hotel_stay['total_stay'], hue=hotel_stay['hotel'])
plt.title("Total no of stay in each hotel")
plt.xlabel('total stay')
plt.ylabel('Count of days')
plt.show()
```

# Which hotel has the long waiting time?

```python
waiting_time = df.groupby('hotel')['days_in_waiting_list'].mean().reset_index()



plt.figure(figsize=(8,5))
sns.barplot (x=waiting_time['hotel'],y=waiting_time['days_in_waiting_list'])
plt.title("Waiting time for each hotel")
plt.xlabel('hotel type')
plt.ylabel('waiting time')
plt.show()
```

# What is the ard across different months?

```python
bookings_months =df.groupby(['arrival_date_month','hotel'])['adr'].mean().reset_index()

months=['january','february','march','april','may','june','july','august','september','october','november','december']

#bookings_months['arrival_date_month']=pd.Categorical(bookings_months['arrival_date_month'],categories=months,ordered=True)

bookings_months=bookings_months.sort_values('arrival_date_month')

bookings_months
```

```python
# setting the chart size

plt.figure(figsize=(15,5))

# Ploting the value in a line chart
sns.lineplot(x=bookings_months['arrival_date_month'],y=bookings_months['adr'],hue=bookings_months['hotel'])
# Seting the label AND title
plt.title('ADR across each months')
plt.xlabel('Month name')
plt.ylabel('ADR')
plt.show()
```

# Which distribution channel has highest adr?

```python
dist_channel_adr=df.groupby(['distribution_channel','hotel'])['adr'].mean().reset_index()

plt.figure(figsize=(15,5))

sns.barplot(x='adr',y='distribution_channel',data=dist_channel_adr,hue='hotel')

plt.title("ADR across each distribution channel")

plt.show()
```
